Importation des Librairies Python

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Chargement du jeu de données 

```python

# Chemin d'accès au fichier CSV
path = "/content/ecoles.csv"

# Charger les données du fichier CSV dans un DataFrame pandas
data = pd.read_csv(path , index_col="code" , decimal=',' , thousands= ' ' )

# Afficher les premières lignes du DataFrame
print(data.head())
```

L'ajout des colonnes des calculs ▶

```python


# Le calcul de la population des moins de 25 ans
data["Population moins de 25 ans"] = data.apply(lambda row: row["Population estimée 2019"] * row["Part des 0 à 24 ans (en %)"] / 100 , axis=1)


# Le calcul de la part de l'enseignement privé
data["Part de l'enseignement privé"] = data.apply(lambda row: row["Total écoles"] / (row["Privé – Maternelles"]+row["Privé – Élémentaires"]) * 100, axis=1)


# le calcul du ratio nombre de jeunes (< 25 ans)/écoles
data["Ratio nombre de jeunes / écoles"] = data.apply(lambda row: row["Part des 0 à 24 ans (en %)"] / row["Total écoles"], axis=1)


#L'ajout des nouvelles colonnes 
data.to_csv(path, index=False)
```

```python
print(data.head())   #Pour afficher la nouvelle base de données 
```

Filtrage de la base de données data ==>


```python
# Méthode 02 pour la création de colonnes et les calculs et la visualisation de données :

#La population de moins de 25 ans: 
data["Population moins de 25 ans11"] = data["Population estimée 2019"] * data["Part des 0 à 24 ans (en %)"] / 100


#La part de l'enseignement privé:
data["Part de l'enseignement privé11"] = data["Total écoles"] / (data["Privé – Maternelles"]+data["Privé – Élémentaires"]) * 100


# le calcul du ratio nombre de jeunes (< 25 ans)/écoles:
data["Ratio nombre de jeunes / écoles11"] = data["Part des 0 à 24 ans (en %)"] / data["Total écoles"]

data.head()
#L'ajout des nouvelles colonnes 
#data.to_csv(path, index=False)
```

La création de la sous_base ==>

```python
#La création de la sous-base de données apartir du jeu de données ecole.csv ==>
data = data.loc[:, ['Département', 'Total écoles', 'Population moins de 25 ans' , "Part de l'enseignement privé" , "Ratio nombre de jeunes / écoles"]]
data.to_csv(path, index=False)

#Affichage de données ==>
data.head()
```
